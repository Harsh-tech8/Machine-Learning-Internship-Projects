```python
threshold = len(df) * 0.4
df_cleaned = df.dropna(thresh=threshold, axis=1)
```

```python
cols_to_drop = ['Street Number Suffix', 'Unit Suffix', 'Structural Notification', 
                'Voluntary Soft-Story Retrofit', 'TIDF Compliance', 'Site Permit']
df_cleaned = df.drop(columns=cols_to_drop, errors='ignore')
```

```python
df_cleaned.info()
```

```python
df.shape
```

```python
cols_to_drop_ = ['Uit', 'Fire Only Permit']
df_cleaned = df_cleaned.drop(columns=cols_to_drop, errors='ignore')
```

```python
date_cols = [
    'Permit Creation Date', 'Current Status Date', 'Filed Date', 
    'Issued Date', 'Completed Date', 'First Construction Document Date', 'Permit Expiration Date'
]
```

```python
for col in date_cols:
    if col in df_cleaned.columns:
        df_cleaned[col] = pd.to_datetime(df_cleaned[col], errors='coerce')
```

```python
text_cols = ['Street Name', 'Current Status', 'Proposed Use', 'Existing Use']
for col in text_cols:
    if col in df_cleaned.columns:
        df_cleaned[col] = df_cleaned[col].astype(str).str.lower().str.strip()
```

```python
num_cols = [
    'Number of Existing Stories', 'Number of Proposed Stories', 
    'Estimated Cost', 'Revised Cost', 'Existing Units', 'Proposed Units'
]
for col in num_cols:
    if col in df_cleaned.columns:
        df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].median())
```

```python
if 'Zipcode' in df_cleaned.columns:
    df_cleaned['Zipcode'] = df_cleaned['Zipcode'].fillna(df_cleaned['Zipcode'].mode()[0])
    df_cleaned['Zipcode'] = df_cleaned['Zipcode'].astype(int).astype(str)
```

```python
df_cleaned = df_cleaned.drop_duplicates(subset=['Record ID'])
```

```python
print("New DataFrame Shape:", df_cleaned.shape)
print("\nRemaining Missing Values per Column:")
print(df_cleaned.isnull().sum())
```

```python
categorical_cols_to_fill = [
    'Street Suffix', 'Description', 'Existing Construction Type Description', 
    'Proposed Construction Type Description', 'Supervisor District', 'Neighborhoods - Analysis Boundaries'
]
```

```python
for col in categorical_cols_to_fill:
    if col in df_cleaned.columns:
        df_cleaned[col] = df_cleaned[col].fillna('Unknown')
```

```python
df_cleaned['Plansets'] = df_cleaned['Plansets'].fillna(0)
df_cleaned['Existing Construction Type'] = df_cleaned['Existing Construction Type'].fillna(-1)
df_cleaned['Proposed Construction Type'] = df_cleaned['Proposed Construction Type'].fillna(-1)
```

```python
df_cleaned['Location'] = df_cleaned['Location'].fillna('Unknown')
```

```python
print(df_cleaned.isnull().sum())
```

```python
df_cleaned['Days_To_Issue'] = (df_cleaned['Issued Date'] - df_cleaned['Filed Date']).dt.days
```

```python
df_cleaned['Days_To_Complete'] = (df_cleaned['Completed Date'] - df_cleaned['Issued Date']).dt.days
```

```python
if 'Filed Date' in df_cleaned.columns:
    df_cleaned['Filed_Year'] = df_cleaned['Filed Date'].dt.year
    df_cleaned['Filed_Month'] = df_cleaned['Filed Date'].dt.month
```

```python
df_cleaned['Cost_Difference'] = df_cleaned['Revised Cost'] - df_cleaned['Estimated Cost']
```

```python
categorical_indices = ['Permit Type', 'Existing Construction Type', 'Proposed Construction Type']
for col in categorical_indices:
    if col in df_cleaned.columns:
        df_cleaned[col] = df_cleaned[col].astype(str)

```

```python
print(df_cleaned[['Days_To_Issue', 'Days_To_Complete', 'Filed_Year', 'Cost_Difference']].head())
print("\nFinal Data Shape:", df_cleaned.shape)
```

```python
print(" AVERAGE DELAYS BY PERMIT TYPE ")
permit_delays = df_cleaned.groupby('Permit Type Definition')['Days_To_Issue'].agg(['count', 'mean', 'median']).sort_values(by='mean', ascending=False)
print(permit_delays.head(5))
print("\n" + "="*50 + "\n")

print("AVERAGE BUDGET INCREASE BY NEIGHBORHOOD ")
neighborhood_costs = df_cleaned.groupby('Neighborhoods - Analysis Boundaries')['Cost_Difference'].agg(['count', 'mean', 'max']).sort_values(by='mean', ascending=False)
print(neighborhood_costs.head(5))
print("\n" + "="*50 + "\n")

if 'Number of Existing Stories' in df_cleaned.columns:
    correlation = df_cleaned['Number of Existing Stories'].corr(df_cleaned['Days_To_Issue'])
    print(f" CORRELATION ANALYSIS ")
    print(f"Correlation between 'Number of Existing Stories' and 'Days_To_Issue': {correlation:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df_cleaned['Cost_Diff_Symlog'] = np.sign(df_cleaned['Cost_Difference']) * np.log1p(np.abs(df_cleaned['Cost_Difference']))

plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
sns.histplot(df_cleaned['Cost_Difference'], bins=50, color='crimson', kde=False)
plt.title('Original Cost Difference (Heavily Skewed by Outliers)')
plt.xlabel('Cost Difference ($)')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
sns.histplot(df_cleaned['Cost_Diff_Symlog'], bins=50, color='teal', kde=True)
plt.title('Symmetric Log Transformed Cost Difference')
plt.xlabel('Log Scale Units')
plt.ylabel('Count')

plt.tight_layout()
plt.show()

print("LOG-TRANSFORMED COLUMN STATISTICS")
print(df_cleaned['Cost_Diff_Symlog'].describe())
```

```python
for col in ['Neighborhoods - Analysis Boundaries', 'Permit Type Definition']:
    if col in df_cleaned.columns:
        target_mean = df_cleaned.groupby(col)['Days_To_Issue'].mean()
        
        df_cleaned[f'{col}_Encoded'] = df_cleaned[col].map(target_mean)


features = [
    'Number of Existing Stories', 'Number of Proposed Stories', 
    'Plansets', 'Cost_Diff_Symlog',
    'Neighborhoods - Analysis Boundaries_Encoded', 
    'Permit Type Definition_Encoded'
]

X = df_cleaned[features]
y = df_cleaned['Days_To_Issue']

print("MODEL READY MATRIX")
print("Features shape (X):", X.shape)
print("Target shape (y):", y.shape)
print("\nFirst 5 rows of your predictor features:")
print(X.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
valid_idx = y.dropna().index
X_clean = X.loc[valid_idx]
y_clean = y.loc[valid_idx]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_clean, y_clean, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
print("Training features shape:", X_train_scaled.shape)
print("Testing features shape:", X_test_scaled.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
model = RandomForestRegressor(n_estimators=50, random_state=42, n_jobs=-1)
model.fit(X_train_scaled, y_train)
```

```python
y_perd = model.predict(X_test_scaled)
```

```python
y_perd
```

```python
r2 = r2_score(y_test, y_perd)
```

```python
r2
```

```python
mae = mean_absolute_error(y_test, y_perd)
print("Mean Absolute Error:", round(mae, 2), "days")

plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=y_perd, alpha=0.4, color='dodgerblue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')
plt.title('Actual vs. Predicted Permit Delays')
plt.xlabel('Actual Days to Issue')
plt.ylabel('Predicted Days to Issue')
plt.tight_layout()
plt.show()
```

```python
import joblib

df_cleaned.to_csv('cleaned_permit_data.csv', index=False)

model_artifacts = {
    'model': model,
    'scaler': scaler,
    'features': features
}
joblib.dump(model_artifacts, 'permit_delay_model.pkl')

```
